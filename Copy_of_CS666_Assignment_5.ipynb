**Task 2**: Load the data using pydicom as a 3D volume and then reslice it! [35 Points]

```python
# TODO: Please upload ct.zip using the file panel on the left.
# Then use the following snippet to extract the data.
```

```python
import zipfile
with zipfile.ZipFile('ct.zip', 'r') as zip_ref:
    zip_ref.extractall('.')
```

```python
# 1) Now loop through all the DICOM files and store them in a 3D numpy array.
# Hint: You can either store them in a list first or read the dimensions of a
# single image slice to properly create the 3D numpy array.
# Hint 2: os.listdir(DIR) gives a list of filenames in a directory.
# Hint 2b: This list is not sorted - make sure you sort it.
# Hint 3: The dcmread function loads a single DICOM file.
# Hint 4: You can then use .pixel_array to access the image data.
```

```python
from pydicom import dcmread
```

```python
# TODO: YOUR CODE FOR LOADING THE VOLUME AS A 3D NUMPY ARRAY
import os
# import numpy as np

DIR = "./ct/"
files_list = os.listdir(DIR)
print(files_list)
print("No. of File : ",len(files_list))
```

```python
files_list = sorted(files_list)
print(files_list)
```

```python
files =[dcmread(os.path.join(DIR,files)) for files in files_list]
single_file = files[0].pixel_array
print(single_file)
print('Dimensions : ',single_file.shape)
```

```python
import numpy as np

r = single_file.shape[0]
c = single_file.shape[1]
l = len(files_list)

new3Darray = np.zeros((l, r, c),dtype=np.float32)

print(new3Darray.shape)
```

```python
for i in range(l):
  new3Darray[i, :, :]=(files[i].pixel_array)
print(new3Darray)

print(new3Darray.shape)
     
```

```python
# 2) Now create and show axial, sagittal, and coronal slices from the 3D volume.
# Hint: Please use imshow(XX, cmap='gray') to show the image.
```

```python
# TODO: YOUR CODE FOR AXIAL
import matplotlib.pyplot as plt

# Create and show an axial slice
print(new3Darray[:, :, 110])
axial_slice = new3Darray[:, :,110]
plt.imshow(axial_slice, cmap='gray')
plt.show()
```

```python
# TODO: YOUR CODE FOR SAGITTAL
sagittal_slice = new3Darray[:, 110, :]
plt.imshow(sagittal_slice, cmap='gray')
plt.show()
```

```python
# TODO: YOUR CODE FOR CORONAL
coronal_slice = new3Darray[110, : , :]
plt.imshow(coronal_slice, cmap='gray')
plt.show()
```

```python
# 1) Please load the Window/Level values from the DICOM file,
# print these values, and then visualize one slice with window/level adjustment.
# Hint: The DICOM header has the following tags.
#   (0028, 1050) Window Center 
#   (0028, 1051) Window Width
# Hint 2: You can use slice[key].value to access DICOM tag values.
# Hint 3: (0028, 1052) Rescale Intercept might be important.
```

```python
print(files[0])
```

```python
# TODO: YOUR CODE
window = files[0].WindowWidth
level = files[0].WindowCenter
rescale = files[0].RescaleIntercept

print("Window : ",window)
print("Level : ",level)
print("Rescale : ",rescale)
```

```python
vmin = level - window/2
vmax = level + window/2
print(vmin)
print(vmax)
plt.imshow(axial_slice + rescale, cmap='gray', vmin=vmin, vmax=vmax)
plt.show()
```

```python
# Similar to Window/Level adjustment for visualization, we can threshold
# the volume to highlight the following components using the Hounsfield Units:
# 1) Fat
# 2) Soft Tissue
# 3) Bones
#
# Please create 3 segmentation masks for these structures.
# Then, please visualize each 3 slices per structure to showcase the segmentation.
# Hint: As a reminder, the following code allows thresholding of a numpy array.
#   new_mask = imagevolume.copy()
#   new_mask[new_mask < XXX] = 0
# Hint2: You might need to cast new_mask to int16 not uint16.
```

```python
imagevolume = new3Darray.astype(np.int16)
```

```python
# TODO: YOUR CODE TO SEGMENT FAT
fat_mask = imagevolume.copy()
fat_mask[fat_mask < -30] = 0
fat_mask[fat_mask > -190] = 0
fat_mask[fat_mask >= -30] = 1
```

```python
# TODO: YOUR CODE TO SEGMENT SOFT TISSUE
soft_tissue_mask = imagevolume.copy()
soft_tissue_mask[soft_tissue_mask < 0] = 0
soft_tissue_mask[soft_tissue_mask >= 100] = 0
soft_tissue_mask[soft_tissue_mask > 0] = 1
```

```python
# TODO: YOUR CODE TO SEGMENT BONES
bone_mask = imagevolume.copy()
bone_mask[bone_mask < 700] = 0
bone_mask[bone_mask >= 3000] = 0
bone_mask[bone_mask > 0] = 1
```

```python
# Are the segmentations good?
```

```python
# TODO: The segementations aren't good enough to show the various components.
```

```python
import matplotlib.pyplot as plt

# plot 3 slices for each structure
slices = [-100, 100, 300]
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10, 10))

for i, slice_idx in enumerate(slices):
    axes[i, 0].imshow(fat_mask[:, :, slice_idx], cmap='gray')
    axes[i, 0].set_title(f'Fat - Slice {slice_idx}')

    axes[i, 1].imshow(soft_tissue_mask[:, :, slice_idx], cmap='gray')
    axes[i, 1].set_title(f'Soft Tissue - Slice {slice_idx}')

    axes[i, 2].imshow(bone_mask[:, :, slice_idx], cmap='gray')
    axes[i, 2].set_title(f'Bones - Slice {slice_idx}')

plt.tight_layout()
plt.show()
```
